```python
# Convert starttime & stoptime from string to date times
data['starttime'] = data['starttime'].apply(dateutil.parser.parse, dayfirst=False)
data['stoptime'] = data['stoptime'].apply(dateutil.parser.parse, dayfirst=False)
data.dtypes
```

```python
gdata_df.insert(0, 'trip_id', range(1, 1 + len(gdata_df)))
gdata_df.head()
```

```python
gdata_df.set_index('trip_id')
```

```python
cols = list(gdata_df.columns.values)
print(cols)
```

```python
gdata_df = gdata_df[['trip_id', 'starttime', 'stoptime', 'tripduration', 'birth year', 'gender', 'bikeid', 'usertype', 'start station id', 'start station name', 'start station latitude', 'start station longitude', 'end station id', 'end station name', 'end station latitude', 'end station longitude']]
```

```python
gdata_df.head()
```

```python
gdata_df.count()
```

```python
spring_df = gdata_df[(gdata_df['starttime'] > '2018-03-01 00:00:00') & (gdata_df['starttime'] < '2018-06-01 00:00:00')]
spring_df.count()
```

```python
summer_df = gdata_df[(gdata_df['starttime'] > '2018-06-01 00:00:00') & (gdata_df['starttime'] < '2018-09-01 00:00:00')]
summer_df.count()
```

```python
fall_df = gdata_df[(gdata_df['starttime'] > '2018-09-01 00:00:00') & (gdata_df['starttime'] < '2018-12-01 00:00:00')]
fall_df.count()
```

```python
winter1_df = gdata_df[(gdata_df['starttime'] > '2018-01-01 00:00:00') & (gdata_df['starttime'] < '2018-03-01 00:00:00')]
winter1_df.count()
```

```python
winter2_df = gdata_df[(gdata_df['starttime'] > '2018-12-01 00:00:00') & (gdata_df['starttime'] < '2018-12-31 23:59:59')]
winter2_df.count()
```

```python
winter_df = winter1_df.append(winter2_df)
```

```python
winter_df.count()
```

```python
spring_df.to_csv('Output/bikes_spring.csv', index=True, header=True)
summer_df.to_csv('Output/bikes_summer.csv', index=True, header=True)
fall_df.to_csv('Output/bikes_fall.csv', index=True, header=True)
winter_df.to_csv('Output/bikes_winter.csv', index=True, header=True)
```
