```python
import pandas as pd
from glob import glob
```

```python
#Read all csv files in Resources folder 
nyBike_files = glob("Resources/*.csv") 
```

```python
#combine in one file
nyBike_df_list = []
for filename in sorted(nyBike_files):
    nyBike_df_list.append(pd.read_csv(filename))

    nyBike2018_df = pd.concat(nyBike_df_list)
```

```python
# save the final file in same/different directory:
nyBike2018_df.to_csv("Resources/merged_nyBike2018.csv", index=False)
```
